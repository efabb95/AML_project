```
import matplotlib.pyplot as plt
%matplotlib inline

def plot_history(network_history):
    plt.figure()
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.plot(network_history.history['loss'])
    plt.plot(network_history.history['val_loss'])
    plt.legend(['Training', 'Validation'])

    plt.figure()
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.plot(network_history.history['acc'])
    plt.plot(network_history.history['val_acc'])
    plt.legend(['Training', 'Validation'], loc='lower right')
    plt.show()
```

```
def count_img_train(cl):
    mypath = path+cl+"/train/*.jpg"
    #path = "./" + mypath
    #print(path)
    count_img=0
    for image in glob.glob(mypath):
        count_img +=1
    #print(count_img)
    return count_img
    
def count_img_val(cl):
    mypath = path+cl+"/val/*.jpg"
    #path = "./" + mypath
    #print(path)
    count_img=0
    for image in glob.glob(mypath):
        count_img +=1
    return(count_img)
```
